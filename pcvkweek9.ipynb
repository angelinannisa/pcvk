7. Make the Opening operation and show the results on the image of the Structuring
Element shaped 7 x 7 square with and without using the OpenCV morphology library,
thus generating the results shows in the figure (Use image “j.png”)

```python
# Define the structuring element (square 7x7)
structuring_element = np.ones((7, 7), dtype=np.uint8)

# Opening without OpenCV
def custom_opening(image, structuring_element):
    erosion = cv2.erode(image, structuring_element)
    opening_result = cv2.dilate(erosion, structuring_element)
    return opening_result

# Opening using OpenCV
opening_result_opencv = cv2.morphologyEx(image, cv2.MORPH_OPEN, structuring_element)

# Opening without OpenCV
opening_result_custom = custom_opening(image, structuring_element)

# Display the original image and opening results
plt.figure(figsize=(8, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Awal')

plt.subplot(1, 3, 2)
plt.imshow(opening_result_opencv, cmap='gray')
plt.title('Citra Hasil Opening')

plt.tight_layout()
plt.show()
```

8. Make the Closing operation and show the results on the image of the Structuring
Element shaped 7 x 7 square with and without using the OpenCV morphology library,
thus generating the results shows in the figure (Use image “j.png”)

```python
# Define the structuring element (square 7x7)
structuring_element = np.ones((7, 7), dtype=np.uint8)

# Closing without OpenCV
def custom_closing(image, structuring_element):
    dilation = cv2.dilate(image, structuring_element)
    closing_result = cv2.erode(dilation, structuring_element)
    return closing_result

# Closing using OpenCV
closing_result_opencv = cv2.morphologyEx(image, cv2.MORPH_CLOSE, structuring_element)

# Closing without OpenCV
closing_result_custom = custom_closing(image, structuring_element)

# Display the original image and closing results
plt.figure(figsize=(8, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Awal')


plt.subplot(1, 3, 2)
plt.imshow(closing_result_opencv, cmap='gray')
plt.title('Citra Hasil Closing')

plt.tight_layout()
plt.show()
```
